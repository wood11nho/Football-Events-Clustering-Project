```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder, StandardScaler
```

```python
# Load data
events = pd.read_csv('data/events.csv')
events
```

```python
games_info = pd.read_csv('data/ginf.csv')
games_info
```

```python
# Filter relevant columns
relevant_columns = ['event_team','event_type', 'event_type2','location', 'assist_method', 'situation', 'bodypart']
events = events[relevant_columns]
events
```

```python
# Drop rows with missing critical values
critical_columns = ['event_team', 'event_type']
events = events.dropna(subset=critical_columns)
events
```

```python
# One-hot encode categorical columns
categorical_columns = ['event_type', 'event_type2', 'location', 'assist_method', 'situation', 'bodypart']
encoder = OneHotEncoder(sparse_output=False)
encoded = encoder.fit_transform(events[categorical_columns])
```

```python
# Create a DataFrame with the encoded features
encoded_feature_names = encoder.get_feature_names_out(categorical_columns)
encoded_features = pd.DataFrame(encoded, columns=encoded_feature_names)
encoded_features
```

```python
# Combine encoded features with the team column
data_encoded = pd.concat([events['event_team'].reset_index(drop=True), encoded_features], axis=1)
data_encoded
```

```python
# Group by team
team_aggregated = data_encoded.groupby('event_team').sum().reset_index()
team_aggregated
```

```python
# Print columns names
team_aggregated.columns
```

```python
# Drop event_type2_nan, location_nan, situation_nan, bodypart_nan
team_aggregated = team_aggregated.drop(columns=['event_type2_nan', 'location_nan', 'situation_nan', 'bodypart_nan'])
team_aggregated
```

```python
team_aggregated.columns
```

```python
# Drop location_19.0 because it is "Not recorded"
team_aggregated = team_aggregated.drop(columns=['location_19.0'])
```

```python
team_aggregated
```

```python
team_aggregated.columns
```

```python
# Count games played for each team
# Combine home and away counts
games_played = games_info['ht'].value_counts() + games_info['at'].value_counts()
team_aggregated['games_played'] = team_aggregated['event_team'].map(games_played).fillna(1)  # Avoid division by zero
```

```python
# Normalize event-related columns by games played
columns_to_normalize = [
    col for col in team_aggregated.columns if col not in ['event_team', 'games_played']
]
for col in columns_to_normalize:
    team_aggregated[col] = team_aggregated[col] / team_aggregated['games_played']
```

```python
team_aggregated
```

```python
# Add aggregate statistics for each team

# Identify event-related columns (e.g., event_type columns)
event_type_columns = [col for col in columns_to_normalize if col.startswith('event_type_')]

# Calculate total_events as the sum of event type columns (normalized by matches)
team_aggregated['total_events'] = team_aggregated[event_type_columns].sum(axis=1)

# Calculate percentage contributions for each event type
for col in columns_to_normalize:
    team_aggregated[f'percent_{col}'] = team_aggregated[col] / team_aggregated['total_events']
```

```python
team_aggregated
```

```python
team_aggregated.columns
```

```python
from itertools import product

# Define interaction categories

event_type_columns = [col for col in team_aggregated.columns if col.startswith('event_type_')]
location_columns = [col for col in team_aggregated.columns if col.startswith('location_')]
assist_method_columns = [col for col in team_aggregated.columns if col.startswith('assist_method_')]
bodypart_columns = [col for col in team_aggregated.columns if col.startswith('bodypart_')]
situation_columns = [col for col in team_aggregated.columns if col.startswith('situation_')]

# Define meaningful combinations
interaction_pairs = [
    (event_type_columns, location_columns),       # Event Type x Location 
    # Why: Different event types often occur in specific areas of the field. For example:
    # "Corner" in "Left wing" or "Penalty conceded" at "Penalty spot".
    # Example pairs:
    # event_type_1 x location_3.0 (Attempt x Centre of the box)
    # event_type_2 x location_4.0 (Corner x Left wing)
    
    (event_type_columns, assist_method_columns),  # Event Type x Assist Method
    # Why: Certain assist methods are more common for specific event types. For example:
    # "Through ball" for "Attempt" or "Cross" for "Corner".
    # Example pairs:
    # event_type_1 x assist_method_4 (Attempt x Through ball)
    # event_type_8 x assist_method_2 (Free kick won x Cross)
    
    (location_columns, bodypart_columns),         # Location x Body Part
    # Why: Certain body parts might be used in specific field locations. For example:
    # "Head" in "Centre of the box" or "Right foot" in "Penalty spot".
    # Example pairs:
    # location_3.0 x bodypart_3.0 (Centre of the box x Head)
    # location_14.0 x bodypart_1.0 (Penalty spot x Right foot)
    
    (location_columns, assist_method_columns),    # Location x Assist Method
    # Why: Assist methods might depend on the field location. For example:
    # "Cross" from "Left wing" or "Pass" in "Centre of the box".
    # Example pairs:
    # location_4.0 x assist_method_2 (Left wing x Cross)
    # location_3.0 x assist_method_1 (Centre of the box x Pass)
    
]

# Generate and calculate interaction terms
for category1, category2 in interaction_pairs:
    for col1, col2 in product(category1, category2):
        team_aggregated[f'{col1}_x_{col2}'] = team_aggregated[col1] * team_aggregated[col2]
```

```python
team_aggregated.head()
```

```python
# Drop the games_played column after normalization
team_aggregated = team_aggregated.drop(columns=['games_played'])
```

```python
# Step 11: Normalize features overall using StandardScaler
scaler = StandardScaler()
features_to_scale = [col for col in team_aggregated.columns if col not in ['event_team']]
scaled_features = scaler.fit_transform(team_aggregated[features_to_scale])
```

```python
# Create final scaled DataFrame
final_team_cluster_data = pd.DataFrame(scaled_features, columns=features_to_scale)
final_team_cluster_data['event_team'] = team_aggregated['event_team']
```

```python
# Step 12: Save processed data for clustering
output_file = 'data/final_team_cluster_data.csv'
final_team_cluster_data.to_csv(output_file, index=False)
print(f"Processed and normalized dataset saved to {output_file}")
```
